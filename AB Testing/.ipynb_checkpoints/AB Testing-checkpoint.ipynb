```python
df_cat= df[['test group', 'converted', 'most ads day', 'most ads hour']]
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05

for variable in df_cat.columns:
    if variable != 'converted':
        # Create a contingency table
        contingency_table = pd.crosstab(df_cat[variable], df_cat['converted'])
        
        # Perform chi-squared test
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        
        # Display results
        print(f"\nChi-squared test for {variable} vs. converted:")
        print(f"Chi-squared value: {chi2:.4f}")
        print(f"p-value: {p:.4f}")
        
        # Check for statistical significance
        if p < alpha:
            print(f" The difference in conversion rates across '{variable}' is **statistically significant**.")
        else:
            print(f" There is **no significant difference** in conversion rates across '{variable}'.")
```

```python
from scipy.stats import shapiro, levene

# Step 1: Check assumptions

# Split data into two groups based on 'converted'
converted_true = df[df['converted'] == True]['total ads']
converted_false = df[df['converted'] == False]['total ads']

# Normality check (Shapiro-Wilk Test)
shapiro_stat_true, shapiro_p_value_true = shapiro(converted_true)
shapiro_stat_false, shapiro_p_value_false = shapiro(converted_false)

print(f"Shapiro-Wilk test (converted=True): p-value = {shapiro_p_value_true:.4f}")
print(f"Shapiro-Wilk test (converted=False): p-value = {shapiro_p_value_false:.4f}")

# Variance equality check (Levene's Test)
levene_stat, levene_p_value = levene(converted_true, converted_false)
print(f"Levene's test for equality of variances: p-value = {levene_p_value:.4f}")
```

```python
from scipy.stats import ttest_ind, mannwhitneyu

# Define alpha
alpha = 0.05

# Extract groups again
converted_true = df[df['converted'] == True]['total ads']
converted_false = df[df['converted'] == False]['total ads']

# Choose the right test
if (shapiro_p_value_true > alpha and 
    shapiro_p_value_false > alpha and 
    levene_p_value > alpha):
    
    # Assumptions met  use independent t-test
    t_stat, t_p_value = ttest_ind(converted_true, converted_false)
    print(f"\nIndependent two-sample t-test: p-value = {t_p_value:.4f}")
    
    if t_p_value < alpha:
        print(" The difference in average total ads is statistically significant.")
    else:
        print(" No significant difference in average total ads between groups.")
        
else:
    # Assumptions not met  use Mann-Whitney U test
    u_stat, u_p_value = mannwhitneyu(converted_true, converted_false, alternative='two-sided')
    print(f"\nMann-Whitney U test: p-value = {u_p_value:.4f}")
    
    if u_p_value < alpha:
        print(" The difference in median total ads is statistically significant.")
    else:
        print(" No significant difference in median total ads between groups.")
```
